```python
# This file creates a target distribution plot for the classification version of the House Prices dataset.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the data
df = pd.read_csv('../data/train.csv')
```

```python
from project.src.utils import get_median_price, classify_price

# Step 2: Convert SalePrice to categorical classes
median_price = get_median_price(df)
df['PriceCategory'] = df['SalePrice'].apply(lambda x: classify_price(x, median_price))
```

```python
# Step 3: Create bar plot
category_counts = df['PriceCategory'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color=['#2E86AB', '#A23B72'], alpha=0.8, edgecolor='black')
plt.title('Target Distribution Plot for Classification\n(Sale Price Categories)', fontsize=14, fontweight='bold')
plt.xlabel('Price Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks([0, 1], ['Below Median', 'Above Median'])
plt.grid(axis='y', alpha=0.3)

# Add count labels on top of bars
for i, count in enumerate(category_counts.values):
    plt.text(i, count + 10, str(count), ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
```
