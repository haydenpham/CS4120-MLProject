```python
# This file creates a target distribution plot for the classification version of the House Prices dataset.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the data
df = pd.read_csv('../data/train.csv')
```

```python
from project.src.utils import classify_price_text

# Step 2: Convert SalePrice to categorical classes
df['PriceCategory'] = df['SalePrice'].apply(classify_price_text)
```

```python
# Step 3: Create bar plot
plt.figure(figsize=(10, 6))
plt.title('Target Distribution Plot for Classification\n(Sale Price Categories)', fontsize=14, fontweight='bold')
plt.xlabel('Price Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
